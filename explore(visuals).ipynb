```python
# Further exploration into tenure
catcols = telco.select_dtypes(include='int')
for col in catcols:
    sns.boxplot(data=telco, x=telco[col], y='tenure')
    plt.show()
```

### Takeaways:
- Longer Tenure
    - 'partner' = 'Yes'
    - 'dependents' = 'Yes'
    - 'multiple_lines' = 'Yes'
    - 'online_security' = 'Yes'
    - 'online_backup' = 'Yes'
    - 'device_protection' = 'Yes'
    - 'tech_support' = 'Yes'
    - 'streaming_tv' = 'Yes'
    - 'streaming_movies' = 'Yes'
    - 'payment_type' = 'Bank transfer (automatic)' or 'Credit Card (automatic)'
    - 'contract_type' = 'One year' or 'Two year'
    - As 'total_services' increase
    - As 'total_extra_services' increase
- Retention cols
    - 'monthly_charges' approx. 60
- Losing cols
    - 'senior_citizen'
- Churn happens around 15-20 months

<a id='additionalexploration'></a>
# Additional Exploration:
<li><a href='#TableOfContents'>Table of Contents</a></li>

```python
# Longer tenure cols vs. churn
longertenure = [
    'partner',
    'dependents',
    'multiple_lines',
    'online_security',
    'online_backup',
    'device_protection',
    'tech_support',
    'streaming_tv',
    'streaming_movies',
    'payment_type',
    'contract_type',
    'total_services',
    'total_extra_services'
]
for x in longertenure:
    sns.histplot(data=telco, x=telco[x], hue='churn', multiple='dodge')
    plt.show()
```

### Takeaways:
- Less Likely to Churn
    - 'partner' = 'Yes'
    - 'dependents' = 'Yes'
    - 'online_security' = 'Yes', 'No internet service'
    - 'online_backup = 'No internet service'
    - 'device_protection' = 'Yes', 'No internet service'
    - 'tech_support' = 'Yes', 'No internet service'
    - 'streaming_tv' = 'No internet service'
    - 'streaming_movies' = 'No internet service'
    - 'contract_type' = 'One year', 'Two year'
    - 'total_services' = '1', '6', '7', '8', '9'
    - 'total_extra_services' = '0', '3', '4', '5', '6'
- AVOID
    - 'payment_type' = 'Electronic check'
    - 'contract_type' = 'Month-to-month'
    - 'total_services'= '3', '4'
    - 'total_extra_services' = '1', '2'

### Additional Comment:
- Target audience:
    - Couples

- Services Emphasis:
    - Cell Services:
        - Not much difference with/without/multiple
